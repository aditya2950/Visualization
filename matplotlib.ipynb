```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(1,10,200)
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
plt.xlabel("This is x data with 1 to 10")
plt.ylabel("This is y data with sin of x")
plt.title("Sin wave")
plt.show()
```

```python
x1=np.random.rand(50)
x1
```

```python
y1=np.random.rand(50)
y1
```

```python
plt.scatter(x1,y1)
```

```python
plt.plot(x1,y1)
```

```python
colors=np.random.rand(50)
sizes=1000*np.random.rand(50)
plt.scatter(x1,y1, c=colors, s=sizes, alpha=1)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```

```python
a=['a','b','c','d','e']
```

```python
b=np.random.rand(5)
b
```

```python
plt.bar(a,b)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")
```

```python
plt.barh(a,b)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")
```

```python
#plt.plot(a,b, '--g')
plt.figure(figsize=(6,2))
plt.plot(a,b, '--g')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")
plt.grid()
plt.show()
```

```python
sizes=1000*np.random.rand(50)
plt.scatter(x1,y1, c='orange', s=sizes, alpha=1, marker='v')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data shown by using V marker")
```

```python

data=[1,2,1,3,4,2,1,5,6,3,4]
plt.hist(data, color="red")
plt.xlabel('Data of x axis')
plt.ylabel('Frequency of data')
plt.title('Histo graph')
plt.show()
```
